```python
# import các thư viện cần thiết
import pandas as pd
import matplotlib.pyplot as plt
print('Import successfully!')
```

```python
# Đọc dữ liệu
data = pd.read_csv('daily-min-temperatures.csv')
data.info()
```

```python
# Đọc các chỉ số min, max, mean... của cột Temp
data['Temp'].describe()
```

```python
# Vẽ biểu đồ hist
plt.hist(data['Temp'], bins = 45);
```

Giá trị xuất hiện nhiều nhất là trong khoảng 10-11 độ.
Có một số giá trị bất thường ở cả 2 phía. Một số rất ít ngày có nhiệt độ cao bất thường trên 23 độ và thấp bất thường dưới 1 độ.

```python
# Vẽ biểu đồ đường phân tích xu hướng thay đổi nhiệt độ theo thời gian trong nawm 1990.
data['Date'] = pd.to_datetime(data['Date'])
data1 = data[(data['Date'] >= pd.to_datetime('01/01/1990')) & (data['Date'] <= pd.to_datetime('12/31/1990'))]
x = data1['Date']
y = data1['Temp']
plt.plot(x, y)
plt.title('Daily Tempurature')
plt.xlabel('Date')
plt.ylabel('Temp')
plt.show()
# data1
```
